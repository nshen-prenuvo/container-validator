```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Read the predictions CSV file
df = pd.read_csv('predictions/all_predictions.csv')

# Create scatter plot
plt.figure(figsize=(10,8))
plt.scatter(df['true_age'], df['predicted_age'], alpha=0.5)
plt.plot([20, 85], [20, 85], 'r--', label='Identity line') # Add identity line

# Fit linear regression
X = df['true_age'].values.reshape(-1, 1)
y = df['predicted_age'].values
reg = LinearRegression().fit(X, y)

# Plot regression line
X_plot = np.array([20, 85]).reshape(-1, 1)
y_plot = reg.predict(X_plot)
plt.plot(X_plot, y_plot, 'g-', label=f'Linear regression (RÂ² = {reg.score(X,y):.3f})')

plt.xlabel('True Age')
plt.ylabel('Predicted Age')
plt.title('Predicted vs True Age')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Print regression parameters
print(f"Slope (coefficient): {reg.coef_[0]:.3f}")
print(f"Intercept: {reg.intercept_:.3f}")
```
